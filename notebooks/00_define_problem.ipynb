# Define the problem we are trying to solve

### Is this multi-label or multi-class?

it's multi-label

### What loss function will be required?

Binary Cross Entropy Loss (BCEWithLogitsLoss).


### What evaluation metric will you optimize?

First, for our problem false negative NF (miss a toxic comment) is much worse than false positive FP (classify a non-toxic comment as toxic) so recall is pritorized over precision.
Second, positives are rare.
And I will use macro PR-AUC as an evaluation metric. In other words, the score is the average of the individual PRs of each predicted column.
