## Check imbalance in dataset

### Measure Imbalance ratio IR = N_majority / N_minority

```python
labels = train_df.columns[2:]
mask = (train_df[labels] == 0).all(axis=1)
num_non_toxic = len(train_df[mask])
print(f"Number of non-toxic comments: {num_non_toxic}")
print(f"Number of toxic comments: {len(train_df) - num_non_toxic}")
```

```python
IR = num_non_toxic / (len(train_df) - num_non_toxic)
print(f"Imbalance Ratio (IR): {IR:.2f}")
```

```python
for column in labels:
    print(f"{column} percentage:")
    print(f"{len(train_df[train_df[column] == 1]) / len(train_df) * 100:.2f}%")
```
