## Check imbalance in dataset

### Measure Imbalance ratio IR = N_majority / N_minority

```python
labels = train_df.columns[2:]
mask = (train_df[labels] == 0).all(axis=1)
num_non_toxic = len(train_df[mask])
print(f"Number of non-toxic comments: {num_non_toxic}")
print(f"Number of toxic comments: {len(train_df) - num_non_toxic}")
```

```python
IR = num_non_toxic / (len(train_df) - num_non_toxic)
print(f"Imbalance Ratio (IR): {IR:.2f}")
```

```python
for column in labels:
    print(f"{column} percentage:")
    print(f"{len(train_df[train_df[column] == 1]) / len(train_df) * 100:.2f}%")
```

## Look for co-occurrence between labels

```python
# count of co‑occurrences (i.e. number of rows where both cols == 1)
co_matrix = train_df[labels].T.dot(train_df[labels])
print(co_matrix)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(8,6))
sns.heatmap(co_matrix, annot=True, fmt="d", cmap="Reds")
plt.title("Label Co-occurrence Matrix")
plt.show()
```

```python
co_prop = co_matrix / len(train_df)
print(co_prop)
```

## Estimate vocab size

We will do simple normalization:

- lowercase

- remove punctuation

- remove URLs

- strip spaces

```python
from collections import Counter
import re

def tokenize(text):
    text = text.lower()
    tokens = re.findall(r"\b\w+\b", text)
    return tokens
```

```python
texts = train_df["comment_text"]

counter = Counter()

for text in texts:
    tokens = tokenize(text)
    counter.update(tokens)

vocab_size = len(counter)
print("Vocabulary size:", vocab_size)
```

```python
counter.most_common(20)
```

#### Coverage Analysis

Suppose we keep only the top 30k words.

How much text do we cover?

```python
total_tokens = sum(counter.values())

top_k = 30000
top_tokens = sum(freq for _, freq in counter.most_common(top_k))

coverage = top_tokens / total_tokens
print(coverage)
```
