## Load the dataset into a hugging face dataset object

```python
import datasets

dataset = datasets.load_dataset("csv", data_files="../data/train.csv")
dataset
```

```python
dataset['train'][0]
```

```python
split = dataset['train'].train_test_split(test_size=0.2, seed=42)
train_set = split['train']
test_val_set = split['test']
test_val_split = test_val_set.train_test_split(test_size=0.5, seed=42)
test_set = test_val_split['train']
val_set = test_val_split['test']
```

```python
train_set.to_csv("../data/train_set.csv", index=False)
val_set.to_csv("../data/val_set.csv", index=False)
test_set.to_csv("../data/test_set.csv", index=False)
```
